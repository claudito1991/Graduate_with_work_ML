# FINDING BEST MODEL FOR A CLASSIFICATION PROBLEM

Here we've work with a college's students database from Kaggle where the main target it's to find the way to predict wheter or not if the graduate will be placed in a job after college. In this study our aim was to find the best model to predict this and we have choosen as cadidates Linear Regression, KNN, GausianNB.

### 2. Generacion de nuevas variables.

```python
#Variables de tipo ratio (indican cuan por encima o por debajo de la mediana estan)
df["r_Age"] = df.Age/np.percentile(df.Age,50)
df["r_CGPA"] = df.CGPA/np.percentile(df.CGPA,50)

#Luego de iterar, notamos que estamos dividiendo una variable por una constante, asi que termina estando correlacioanda con la variable original.
df.drop(columns=["r_Age","r_CGPA"],inplace=True)
```

```python
#Interaccion de variables y potencias (solo consideramos Age, GPA e Internships por ser las unicas numericas (ordinales en este caso)).
poly = PolynomialFeatures(degree=3,interaction_only=False,include_bias = False)
vars_interactions = poly.fit_transform(df[["Age","CGPA","Internships"]]) #Deberiamos sumar el resto?

cols_names = poly.get_feature_names(["Age","CGPA","Internships"])
cols_names = [x.replace(" ", "_") for x in cols_names] #sin espacios en blanco.
df_inter = pd.DataFrame(vars_interactions, columns = cols_names)
df_inter.drop(columns =["Age","CGPA","Internships"], inplace = True)
```

```python
#Dataframe.
df_vars = pd.concat([df,df_inter],axis = 1)
```

```python
df_vars["Stream"] = [x.replace(" ", "_") for x in df_vars.Stream]
```

```python
df_vars.info()
```

### 3. Preprocesamiento (transformacion a dummys y estandarizcion)

```python
#Solo las variables categoricas
df_vars = pd.get_dummies(df_vars,columns = ["Gender","Stream"])
```

```python
df_vars.info()
```

```python
#No vamos a estandarizar las dummys o las booleanas.
vars_to_standarize = []
for col in df_vars.columns:
    if sum([x in col for x in ["Gender","Stream","Hostel","History","PlacedOr"]]) == 0 : 
        vars_to_standarize.append(col)
        
```

```python
#Estandarizacion ((x-mu_)/sigma_**2) 
scaler = StandardScaler()
df_vars[vars_to_standarize] = scaler.fit_transform(df_vars[vars_to_standarize])
```

```python
dummys_to_exclude = ['Gender_Male','Stream_Information_Technology']
target_var = ['PlacedOrNot']
vars_to_exclude = [] #todos los ratios menos la interaccion Age_CGPA.
vars_to_model = list(set(df_vars.columns) - set(dummys_to_exclude + target_var + vars_to_exclude))
```

```python
df_vars[vars_to_model].info()
```

### 4. Split and train.

```python
X_train, X_test, y_train, y_test = train_test_split(df_vars[vars_to_model], 
                                                    df_vars[target_var], 
                                                    test_size=0.20, 
                                                    random_state=42)
```

```python


def baseline_model(X_train: pd.DataFrame ,
                   y_train: pd.DataFrame) -> pd.DataFrame:
    '''
    Compara el f1_score de los modelos LogReg, GNB (generan probabilidades) y 
    KNN (genera clase a la que pertenece la instancia por votacion, no probabiliad)
    Para ser justos en esta comparativa es necesario para los modelos de probabilidad 
    estimar el mejor cutoff (.5 no necesariamente es el mejor) para el calculo del f1_score.
    En el caso de los modelos que proveen directamente la clase se puede calcular directamente 
    el f1_score.
    F1 = 2 * (precision * recall) / (precision + recall)
    
    :param X_train: conjunto train de variables.
    :param y_train: vector de clase en train.
    :return: DataFrame con las metricas por xval.
    '''
    
    models = [
          ('LogReg', LogisticRegression(max_iter=1000)), 
          ('KNN', KNeighborsClassifier()),
          ('GNB', GaussianNB())
        ]
    
    r = []

   
    kfold = KFold(n_splits=5, shuffle=True, random_state=90210)
    
    for name, model in models:
        print(f"CV model {name}")
        for i_train,i_test in kfold.split(X_train):
            model.fit(X_train.iloc[i_train],y_train.iloc[i_train].values.ravel())
            if name in ('LogReg','GNB'):
                p_pred = model.predict_proba(X_train.iloc[i_test])[:,1]
                fpr, tpr, thresholds = roc_curve(y_train.iloc[i_test], p_pred)
                # calculate the g-mean for each threshold
                gmeans = np.sqrt(tpr * (1-fpr))
                ix = np.argmax(gmeans)
                optimal_threshold = thresholds[ix]
                y_pred = np.where(p_pred >= optimal_threshold,1,0)
            else:
                y_pred = model.predict(X_train.iloc[i_test])
                    
            metric_val = f1_score(y_train.iloc[i_test],y_pred)
            r.append((name,(np.mean(metric_val))))
    
    result = pd.DataFrame(r, columns=["model","metric_mean"])
    
    return result
```

```python
baseline_model = baseline_model(X_train,y_train)
```

```python
baseline_model
```

```python
baseline_model.boxplot(column = "metric_mean", by = "model", figsize = (10,6))
```

Considerando los modelos **GaussianNB**, **KNeighborsClassifier** y **LogisticRegression** el mejor modelo (bajo el esquema de crossvalidacion sobre el conjunto train) considerando la metrica del area bajo la curva ROC corresponde al  KNeighborsClassifier, sin embargo este modelo no brinda mayor informacion sobre la importancia de las variables al momento de clasificar, es por esto que podemos considerar el tradeoff entre explicabilidad y poder predictivo.

#### 3.1 Train baseline model y optimizacion de hiperparametros.

```python
model = LogisticRegression(max_iter = 1000,solver = 'saga')
```

```python
# Definimos la penalizacion y los valores de los parametros.
penalty = ['elasticnet']
l1_ratios = np.round(np.arange(0, 1, 0.1).tolist(),2)
# Definicion del grid.
grid = dict(penalty=penalty ,l1_ratio = l1_ratios)
```

```python
grid
```

```python
cv = KFold(n_splits=5,shuffle=True, random_state=90210)
grid_search = GridSearchCV(estimator=model,
                           param_grid=grid, 
                           n_jobs=-1, 
                           cv=cv, 
                           scoring='roc_auc')
grid_result = grid_search.fit(X_train, y_train.values.ravel())
```

```python
print("Best: %f using %s" % (grid_result.best_score_, grid_result.best_params_))
print("-----------------------------------------")
means = grid_result.cv_results_['mean_test_score']
stds = grid_result.cv_results_['std_test_score']
params = grid_result.cv_results_['params']
for mean, stdev, param in zip(means, stds, params):
    print("%f (%f) with: %r" % (mean, stdev, param))
```

#### El modelo de regresion logistica con penalizacion: elasticnet y parametro l1_ratio=.6 tiene un ROC_AUC en train de  .91 y en test de .89

```python
model.set_params(penalty = "elasticnet",l1_ratio = 0.6)
```

```python
model.fit(X_train,y_train.values.ravel())
```

```python
y_pred = model.predict_proba(X_test)[:,1]
```

```python
# Estimacion del AUC ROC en test.
score_auc = roc_auc_score(y_test, y_pred)
print('Logistic: ROC AUC=%.3f' % (score_auc))
```

```python
# Curva ROC y optimal_threshold
fpr, tpr, thresholds = roc_curve(y_test, y_pred)
gmeans = np.sqrt(tpr * (1-fpr))
ix = np.argmax(gmeans)
optimal_threshold = thresholds[ix]
print('Optimal threshold: %.3f' % (optimal_threshold))
```

```python
# plot the roc curve for the model
plt.plot(fpr, tpr, marker='.', label='Logistic')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
```

```python
#Matriz de confusion con el mejor thresholds.
conf_mat = confusion_matrix(y_test,np.where(y_pred>=optimal_threshold,1,0)) 
conf_mat
```

```python
sns.heatmap(conf_mat, annot=True,fmt='.4g')
```

```python
#f1 score utilizando el mejor cutoff
f1_score(y_test,np.where(y_pred>=optimal_threshold,1,0))
```

Las matrices de confusion que se compararon anteriormente sirven para verificar que al momento de asignar una clase es necesario evaluar el mejor cutoff (punto de corte de probabildiad para pertenecer a una clase u otra) y esto a su vez dependera del problema.
